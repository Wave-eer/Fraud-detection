```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

cc_df = pd.read_csv("../data/creditcard.csv")

cc_df.shape
cc_df.head()
cc_df.info()
cc_df.isnull().sum()
cc_df.duplicated().sum()
cc_df.drop_duplicates(inplace=True)

cc_df["Class"].value_counts()
cc_df["Class"].value_counts(normalize=True) * 100

plt.figure(figsize=(5, 4))
sns.countplot(data=cc_df, x="Class")
plt.title("Fraud Distribution")
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(cc_df["Amount"], bins=50)
plt.title("Transaction Amount Distribution")
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(cc_df["Time"], bins=50)
plt.title("Time Distribution")
plt.show()

plt.figure(figsize=(7, 5))
sns.boxplot(data=cc_df, x="Class", y="Amount")
plt.title("Amount by Class")
plt.show()

plt.figure(figsize=(7, 5))
sns.boxplot(data=cc_df, x="Class", y="Time")
plt.title("Time by Class")
plt.show()

cc_df.groupby("Class")["Amount"].describe()
```
