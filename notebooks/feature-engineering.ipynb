# Feature Engineering Analysis
This notebook documents and visualizes the custom-engineered velocity, sharing, and temporal features added to the e-commerce transaction dataset.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys

# Append src path
sys.path.append('..')
from src.data_preprocessing import load_data, clean_fraud_data, engineer_features
```

## 1. Load and Clean Data
We load and map the geolocation coordinates first, then run our feature engineering logic.


```
fraud_df, ip_df, _ = load_data('../data')
cleaned_fraud = clean_fraud_data(fraud_df, ip_df)
engineered_df = engineer_features(cleaned_fraud)
```

## 2. Visualizing Account Age (`time_since_signup`)
Account age is a crucial signal. Let's see how the duration in hours between account creation and transaction differs between legitimate and fraudulent cases.


```
plt.figure(figsize=(7,5))
sns.boxplot(data=engineered_df, x="class", y="time_since_signup", palette='Set2')
plt.title("Account Age (Hours since Signup) by Class")
plt.xlabel("Class (0=Legit, 1=Fraud)")
plt.ylabel("Time since Signup (Hours)")
plt.show()
```

## 3. Visualizing Device Sharing Counts
We group by `device_id` to count how many users shared the same physical device. Multiple users sharing a device is a strong indication of a botnet or fake accounts creation.


```
plt.figure(figsize=(7,5))
sns.countplot(data=engineered_df, x="device_sharing_count", hue="class", palette='Set3')
plt.title("Device Sharing Counts by Class")
plt.xlabel("Number of Users Sharing the Same Device")
plt.ylabel("Count (log-scale)")
plt.yscale("log")
plt.legend(title="Class (0=Legit, 1=Fraud)")
plt.show()
```

## 4. Visualizing Device Velocity (`device_tx_count_1d`)
This measures the number of transactions processed on the same device in a sliding 24-hour window. Rapid transaction rates indicate high-velocity fraud.


```
plt.figure(figsize=(7,5))
sns.boxplot(data=engineered_df, x="class", y="device_tx_count_1d", palette='Set1')
plt.title("Device 24h Transaction Velocity by Class")
plt.xlabel("Class (0=Legit, 1=Fraud)")
plt.ylabel("Transaction Count in 24h Window")
plt.show()
```

## 5. Visualizing IP Velocity (`ip_tx_count_1d`)
This measures the number of transactions originating from the same IP address in a sliding 24-hour window.


```
plt.figure(figsize=(7,5))
sns.boxplot(data=engineered_df, x="class", y="ip_tx_count_1d", palette='Set2')
plt.title("IP 24h Transaction Velocity by Class")
plt.xlabel("Class (0=Legit, 1=Fraud)")
plt.ylabel("Transaction Count in 24h Window")
plt.show()
```
